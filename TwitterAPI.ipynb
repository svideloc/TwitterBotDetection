### For loop to grab users last 30 tweets & Turns into dataframe df_tweets

```python
N = len(accounts)//30+1
df_tweets = pd.DataFrame(columns = ['id', 'text', 'source', 'user_id', 'truncated', 'in_reply_to_status_id', 'in_reply_to_user_id', 'in_reply_to_screen_name', 'geo', 'place', 'contributors', 'retweet_count', 'favorite_count', 'favorited', 'retweeted', 'created_at'])
twitter_info = {}
for i, chunk in enumerate(np.array_split(accounts, N)):
    for j, u in enumerate(chunk):
        try:
            count = 30
            user = api.user_timeline(user_id = u, count = count)
            time.sleep(2)
            index = 30*i+j
            print(index)
            for tweet in range(count):
                tweet_id = user[tweet].id
                text = user[tweet].text
                source = user[tweet].source
                truncated = user[tweet].truncated
                in_reply_to_status_id = user[tweet].in_reply_to_status_id
                in_reply_to_user_id = user[tweet].in_reply_to_user_id
                in_reply_to_screen_name = user[tweet].in_reply_to_screen_name
                geo = user[tweet].geo
                place = user[tweet].place
                contributors = user[tweet].contributors
                retweet_count = user[tweet].retweet_count
                favorite_count = user[tweet].favorite_count
                favorited = user[tweet].favorited
                retweeted = user[tweet].retweeted
                created_at = user[tweet].created_at
                twitter_info[index] = {
                                   'id': tweet_id, 
                                   'text' : text, 
                                   'source' : source,
                                  'user_id': u,
                                  'truncated':truncated,
                                  'in_reply_to_status_id':in_reply_to_status_id,
                                  'in_reply_to_user_id': in_reply_to_user_id,
                                  'in_reply_to_screen_name':in_reply_to_screen_name,
                                  'geo' :geo,
                                  'place':place,
                                  'contributors':contributors,
                                  'retweet_count':retweet_count,
                                  'favorite_count':favorite_count,
                                  'favorited':favorited,
                                  'retweeted': retweeted,
                                  'created_at': created_at}
                df_tweets_real = pd.DataFrame.from_dict(twitter_info, orient = 'index')
                df_tweets = df_tweets.append(df_tweets_real)
                twitter_info = {}
        except:
            pass
```
